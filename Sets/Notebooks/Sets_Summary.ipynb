*Python sets are unordered collections of unique, immutable elements*

### ðŸ”‘ Key Characteristics
- **Mutable:** You can add, modify, or remove items after creation
- **Unordered collection** - Elements don't maintain insertion order (though Python 3.7+ preserves insertion order)
- **Unique elements only** - Automatically removes duplicates
- **Fast membership testing** - O(1) average time complexity for `in` operations
- **Hashable elements only** - Elements must be immutable (strings, numbers, tuples, frozensets)
- **Mathematical operations** - Support union, intersection, difference operations

### When to use sets
- Removing duplicates from a list
- Membership testing (faster than lists)
- Mathematical set operations
- Tracking unique items
