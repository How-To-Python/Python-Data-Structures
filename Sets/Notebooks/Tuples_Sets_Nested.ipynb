# Sets with Tuples (Nested Structures)
- set can contain tuples, because tuples are hashable

```python
coordinates = {(0, 0), (1, 1), (2, 2), (0, 0)}  # Note duplicate (0, 0)
print(f"Coordinate set: {coordinates}")

# Set of name-age pairs
people = {("Alice", 25), ("Bob", 30), ("Charlie", 35)}
print(f"People set: {people}")

# Mixed tuples
mixed_tuples = {(1, "a"), (2, "b"), (3, "c")}
print(f"Mixed tuples: {mixed_tuples}")
```
