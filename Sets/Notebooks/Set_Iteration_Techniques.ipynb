# Set Iteration Techniques

**Various ways to iterate through sets efficiently**

### Basic Set Iteration

Sets are iterable, but unordered (though insertion order is preserved in Python 3.7+).

```python
# Basic iteration through sets
fruits = {"apple", "banana", "cherry", "date", "elderberry"}

print("Basic iteration:")
for fruit in fruits:
    print(f"  {fruit}")

# Iteration with processing
print("\nProcessed iteration:")
for fruit in fruits:
    print(f"  {fruit.upper()} has {len(fruit)} letters")
```

### Advanced: Working with Tuples in Sets
**Compare tuple elements**

```python
# Your data sets
set1 = {('112384', '102'), ('109830', '705'), ('112384', '200'), ('109830', '641'), ('112384', '641'), ('109830', '398')}
set2 = {('104257', '197'), ('104257', '129'), ('104257', '102')}

print(f"Set 1: {set1}")
print(f"Set 2: {set2}")


# Extract second items from each set
for set1_item in set1:
    set1_actor_id = set1_item[1]
    for set2_item in set2:
        set2_actor_id = set2_item[1]
        if set1_actor_id == set2_actor_id:
            print(f"Matching actor ID found: {set1_actor_id}")
            print(f"From Set 1: {set1_item} and From Set 2: {set2_item}")
```
