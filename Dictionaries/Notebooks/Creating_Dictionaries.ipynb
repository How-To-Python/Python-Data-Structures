### Create empty dictionaries

```python
empty_dict = {}
also_empty = dict()

print("Empty with {}:", empty_dict)
print("Empty with dict():", also_empty)
print("Both are equal:", empty_dict == also_empty)
```

### Create Dictionary With Values

#### Using Curly Braces

```python
# Create a dictionary with initial values
person = {
    "name": "John",
    "age": 30,
    "city": "New York"
}
print("Person:", person)
print("Name:", person["name"])
print("Age:", person["age"])
```

```python
# Create list of dictionaries
people = [{"name": "Alice", "age": 25}, {"name": "Bob", "age": 30},{"name": "Charlie", "age": 35}]
print("People List of Dictionaries:", people)
print("First Person:", people[0])
print("First Person's Name:", people[0]["name"])
print("Second Person's Age:", people[1]["age"])
print("Third Person's Name:", people[2]["name"])
print("Third Person's Age:", people[2]["age"])  
```

#### Using `dict()` constructor
- `dict()` can create dictionaries in several ways
    - Keyword arguments
    - List of tuples
    - Zip two lists

```python
# Using `dict()` constructor with keyword arguments
student = dict(name="Rita", age=20, major="CS")
print("Student:", student)
print("Name:", student["name"])
print("Age:", student["age"])
print("Major:", student["major"])
```

```python
# Using `dict()` constructor with list of tuples
colors = dict([("red", "#FF0000"), ("green", "#00FF00"), ("blue", "#0000FF")])
print("Colors:", colors)
print("Red Hex:", colors["red"])
print("Green Hex:", colors["green"])
print("Blue Hex:", colors["blue"])
 
```

```python
# Using `dict()` constructor with zip two lists
# this will create a dictionary from two lists: one for keys and one for values
keys = ["apple", "banana", "cherry"]
values = [1.20, 0.50, 2.00]
fruits = dict(zip(keys, values))
print("Fruits:", fruits)
```
