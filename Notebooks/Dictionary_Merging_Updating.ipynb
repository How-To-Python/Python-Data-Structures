# Dictionary Merging and Updating
- `update()` method for merging dictionaries
- merging with `**` operator (Python 3.5+)

### `update()` Method
- `update()` method with another dictionary
- `update()` with keyword arguments
- `update()` with list of tuples

```python
student = {"name": "Rita", "age": 20, "major": "CS"}
# adding single item using update() method
student.update({"advisor": "Dr. Smith"})
print("After adding advisor:", student)

# adding multiple items using update() method with another dictionary
additional_info = {"graduation_year": 2024, "minor": "Math"}
student.update(additional_info)
print("After adding additional info:", student)

# adding multiple items using update() method with keyword arguments(kwargs)
student.update(honors=True, scholarship="Full Ride")
print("After adding honors and scholarship:", student)

# adding multiple items using update() method with list of tuples
student.update([("internship", "Google"), ("club", "Coding Club")])
print("After adding internship and club:", student)
```

```python
# update() method for merging dictionaries
base_config = {
    "host": "localhost",
    "port": 8080,
    "debug": False
}

user_config = {
    "port": 3000,
    "debug": True,
    "api_key": "secret123"
}

print("Base config:", base_config)
print("User config:", user_config)

# Update base with user settings
base_config.update(user_config)
print("\nAfter update():", base_config)
```

```python
# Different ways to use update()
settings = {"theme": "light", "lang": "en"}

print("Initial settings:", settings)

# Update with dictionary
settings.update({"theme": "dark", "font_size": 14})
print("After dict update:", settings)

# Update with keyword arguments
settings.update(auto_save=True, notifications=False)
print("After kwargs update:", settings)

# Update with list of tuples
settings.update([("version", "1.0"), ("author", "Alice")])
print("After tuples update:", settings)
```

### Dictionary merging with `**` operator (Python 3.5+): Dictionary Upacking
- dictionary unpacking is done using the `**` operator, it unpacks the key-value pairs from each dictionary
- use `**` to unpack a dictionary into keyword arguments
- the keys and values are upacked together
- [More about Upacking Iterables](http://github.com/How-To-Python/Python/blob/main/Notebooks/UnpackingIterables.ipynb)

```python
# Modern merging with ** operator (Python 3.5+)
dict1 = {"a": 1, "b": 2}
dict2 = {"c": 3, "d": 4}
dict3 = {"b": 20, "e": 5}  # 'b' will override dict1's value

# Merge multiple dictionaries
merged = {**dict1, **dict2, **dict3}
print("Dict1:", dict1)
print("Dict2:", dict2)
print("Dict3:", dict3)
print("Merged:", merged)

# Order matters - later values override earlier ones
merged_different_order = {**dict3, **dict1, **dict2}
print("Different order:", merged_different_order)
```
